## General Relativistic Hydrodynamics (GRHD)
---
The _evolution of matter_ can be described by the **conservation equations** of the matter current, $J^{\mu}$, and energy-momentum, $T^{\mu \nu}$, as,
\begin{align*}
    \nabla_{\mu} J^{\mu}     \ &= \ 0   \\
    \nabla_{\mu} T^{\mu \nu} \ &= \ 0 .
\end{align*}
We consider a **perfect fluid** with matter current and energy-momentum tensor defined as,
\begin{align*}
    J^{\mu}
    \ &\equiv \
    \rho u^{\mu}  \\
    T^{\mu\nu}
    \ &\equiv \
    \rho h u^{\mu} u^{\nu} \ + \ p g^{\mu \nu} 
\end{align*}
in which, $\rho$, is the rest-mass density, $u^{\mu}$, is the fluid four-velocity, normalised such that, $g_{\mu\nu}u^{\mu} u^{\nu} = -1$, the enthalpy is defined as,
\begin{equation*}
    h
    \ \equiv \
    1 \ + \ \epsilon \ + \ p/\rho 
\end{equation*}
with, $\epsilon$, the specific internal energy and, $p$, the pressure.

The system of evolution equations needs to be closed with an **equation of state (EOS)**,
\begin{align*}
    p \ = \ p \left(\rho, \epsilon \right)
\end{align*}
that relates the pressure to the density and internal energy.

Recall that the projector in the time direction is, $n_{\mu}$, while the projector on a time-slice (the orthogonal complement of $n_{\mu}$), is given by,
\begin{align*}
    \perp^{\mu}_{\ \ \nu} \ \equiv \ \delta^{\mu}_{\ \ \nu} + n^{\mu} n_{\nu} .
\end{align*}

The energy-momentum tensor can be decomposed as,
\begin{align*}
T_{\mu\nu}
\ = \
E \, n_{\mu} n_{\nu}
\, + \,
S_{\mu} n_{\nu}
\, + \,
S_{\nu} n_{\mu}
\, + \,
S_{\mu \nu}
\end{align*}
where the different projections are given by,
\begin{align*}
    E
    \ &\equiv \ \ \ \
    n_{\mu} \ \ n_{\nu} \ \ \, T^{\mu \nu} 
    \ = \
    \rho h W^{2} - p 
    \\
    S^{i} 
    \ &\equiv \
    - n_{\mu} \perp^{i}_{\ \ \nu} T^{\mu \nu}
    \ = \
    \rho h W^{2} v^{i} \\
    S^{ij} \
    &\equiv \ \ \ \ \,
    \perp^{i}_{\ \ \mu} \perp^{j}_{\ \ \nu} T^{\mu \nu}
    \ = \
    \rho h W^{2} v^{i} v^{j} + p \gamma^{ij} .
\end{align*}

$S \equiv \gamma^{ij} S_{ij} = \rho h W^{2} v^{i} v_{i} + 3 p$

Note that (for a perfect fluid),
\begin{align*}
S^{i}  \ &= \ (E + p) v^{i} \\
S^{ij} \ &= \ S^{i} v^{j} + p \gamma^{ij}
\end{align*}



We can 

\begin{align*}
    W \ &\equiv \ -n_{\mu} u^{\mu} \\
    W v^{i}
    \ &\equiv \ \ \perp^{i}_{\ \ \nu} u^{\nu}
    \ = \ u^{i}  - W n^{i}
\end{align*}



Define the three-velocity of the fluid as measured by a Eulerian observer (i.e. fixed to the coordinate frame),
\begin{align*}
    W
    \ &= \
    \alpha u^{0} \\
    v^{i}
    \ &= \
    \frac{1}{\alpha}
    \left(
    \frac{u^{i}}{u^{0}}
    \ + \
    \beta^{i}
    \right)
\end{align*}
and since, $g_{\mu\nu} u^{\mu} u^{\nu} = -1$, we also find that,
\begin{align*}
    W
    \ = \
    \frac{1}{\sqrt{1-\gamma_{ij} v^{i}v^{j}}} ,
\end{align*}
which is the definition of the Lorenz factor in special relativity.

The state of the matter can thus be summerized by the five **primitive variables**, $\boldsymbol{P} \equiv (\rho, v_{i}, \epsilon)$.



The **conservation of matter current**, $\nabla_{\mu} J^{\mu} = 0$, can be written as,
\begin{align*}
\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \rho u^{\mu} \right) \ = \ 0 .
\end{align*}
Using that, $\sqrt{-g} = \alpha \sqrt{\gamma}$, we find,
\begin{align*}
\partial_{t} \left( \alpha \sqrt{\gamma} \rho u^{t} \right) \ + \
\partial_{i} \left( \alpha \sqrt{\gamma} \rho u^{i} \right)
\ = \ 0
\end{align*}
Using that, $W = \alpha u^{t}$, and that, $u^{i} = W \left(v^{i} + n^{i}\right)$, we find,
\begin{align*}
\partial_{t} \left( \sqrt{\gamma} \rho W \right) \ + \
\partial_{i} \left( \alpha \sqrt{\gamma} \rho W \left(v^{i} + n^{i}\right) \right)
\ = \ 0
\end{align*}
Defining, $D \equiv \rho W$, and using that, $n_{\mu} = (-\alpha, 0)$, and thus, $n^{\mu} = (1/\alpha, -\beta^{i}/\alpha)$,
\begin{align*}
\partial_{t} \left( \sqrt{\gamma} D \right) \ + \
\partial_{i} \left( \sqrt{\gamma} D (\alpha v^{i} - \beta^{i}) \right)
\ = \ 0
\end{align*}

The **conservation of momentum**, $\perp_{i \nu}\nabla_{\mu} T^{\mu\nu} = 0$, or the projection of energy-momentum conservation on the time-slice, and the **conservation of energy**, $n_{\nu}\nabla_{\mu} T^{\mu\nu} = 0$, or the projection of energy-momentum conservation in the time-direction, can be rewritten in a conservtive form.

First note that, using the derivative of a product rule,
\begin{align*}
\perp_{i \nu}\nabla_{\mu} T^{\mu\nu}
\ &= \
\nabla_{\mu} \left( \perp_{i \nu} T^{\mu\nu} \right)
\, - \,
T^{\mu\nu} \, \nabla_{\mu} \perp_{i \nu}
\end{align*}
and,
\begin{align*}
n_{\nu}\nabla_{\mu} T^{\mu\nu}
\ &= \
\nabla_{\mu} \left( n_{\nu} \, T^{\mu\nu} \right)
\, - \,
T^{\mu\nu} \, \nabla_{\mu} n_{\nu} 
\end{align*}


Now, since, $\perp_{i \nu} n^{\nu} = 0$, and since, $S^{\mu}$, and, $S^{\mu \nu}$, are already projections on the time-slice, we have,
\begin{align*}
\perp_{i \nu} T^{\mu\nu}
\ &= \
\perp_{i \nu}
\left(
E \, n^{\mu} n^{\nu} + S^{\mu} n^{\nu} + S^{\nu} n^{\mu} + S^{\mu \nu}
\right) \\
\ &= \
S_{i} n^{\mu} + S^{\mu}_{\ \ i}
\end{align*}
and
\begin{align*}
n_{\nu} T^{\mu\nu}
\ &= \
n_{\nu}
\left(
E \, n^{\mu} n^{\nu} + S^{\mu} n^{\nu} + S^{\nu} n^{\mu} + S^{\mu \nu}
\right) \\
\ &= \
-E \, n^{\mu} - S^{\mu}
\end{align*}

Furthermore, since, $n^{\mu} n_{\mu} = -1$, we have, $n^{\mu} \nabla_{\nu} n_{\mu} = 0$, and since, $\nabla_{\mu} g_{\rho \sigma} = 0$, we have that, $\nabla_{\mu} \perp_{i \nu} = n_{i} \nabla_{\mu} n_{\nu} + n_{\nu} \nabla_{\mu} n_{i}$, using this, we find,
\begin{align*}
T^{\mu\nu} \, \nabla_{\mu} \perp_{i \nu} 
\ &= \ \left(
E \, n^{\mu} n^{\nu} + S^{\mu} n^{\nu} + S^{\nu} n^{\mu} + S^{\mu \nu}
\right) \left( n_{i} \nabla_{\mu} n_{\nu} + n_{\nu} \nabla_{\mu} n_{i} \right) \\
\ &= \ S^{\nu} n_{i} a_{\nu} - n_{i} S^{jk} K_{jk} - E a_{i} + S^{j} K_{ji} \\
\ &= \ - E a_{i} + S^{j} K_{ji}
\end{align*}
where, in the last equality, we used that, $n_{i}=0$, for spatial indices $i$.
\begin{align*}
T^{\mu\nu} \, \nabla_{\mu} n_{\nu} 
\ &= \ \left(
E \, n^{\mu} n^{\nu} + S^{\mu} n^{\nu} + S^{\nu} n^{\mu} + S^{\mu \nu}
\right) \nabla_{\mu} n_{\nu} \\
\ &= \ S^{\nu} a_{\nu} - S^{jk} K_{jk}
\end{align*}
where we used the definitions,
$K_{\alpha \beta} \equiv - \perp^{\mu}_{\ \ \alpha} \nabla_{\mu} n_{\beta}$
and
$a_{\mu} \equiv n^{\mu} \nabla_{\mu} n^{\nu}$,
and the fact that, $S^{\mu} = \perp^{\mu}_{\ \ j} S^{j}$, and, $S^{\mu \nu} = \perp^{\mu}_{\ \ j} \perp^{\nu}_{\ \ k} S^{j k}$.

\begin{align*}
\nabla_{\mu} \left( \perp_{i \nu} T^{\mu\nu} \right)
\ &= \
\partial_{\mu} \left( \perp_{i \nu} T^{\mu\nu} \right)
+ \Gamma^{\mu}_{\ \ \mu \rho} \perp_{i \nu} T^{\rho\nu}
- \Gamma^{\rho}_{\ \ \mu i} \perp_{\rho \nu} T^{\mu\nu}  \\
\ &= \
\partial_{\mu} \left( \perp_{i \nu} T^{\mu\nu} \right)
+ \frac{1}{\sqrt{-g}} \, \partial_{\mu} \sqrt{-g} \perp_{i \nu} T^{\mu\nu}
- \Gamma^{\rho}_{\ \ \mu i} \perp_{\rho \nu} T^{\mu\nu}  \\
\ &= \
\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \perp_{i \nu} T^{\mu\nu} \right) - \Gamma^{\rho}_{\ \ \mu i} \perp_{\rho \nu} T^{\mu\nu} \\
\ &= \
\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \left( S_{i} n^{\mu} + S^{\mu}_{\ \ i} \right) \right) - \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
\end{align*}

\begin{align*}
\nabla_{\mu} \left( n_{\nu} T^{\mu\nu} \right)
\ &= \
\partial_{\mu} \left( n_{\nu} T^{\mu\nu} \right)
+ \Gamma^{\mu}_{\ \ \mu \rho} n_{\nu} T^{\rho\nu}  \\
\ &= \
\partial_{\mu} \left( n_{\nu} T^{\mu\nu} \right)
+ \frac{1}{\sqrt{-g}} \, \partial_{\mu} \sqrt{-g} \, n_{\nu} T^{\mu\nu}  \\
\ &= \
\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \, n_{\nu} T^{\mu\nu} \right) \\
\ &= \
-\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \left( E \, n^{\mu} + S^{\mu} \right) \right)
\end{align*}

\begin{align*}
0 \ &= \
\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \left( S_{i} n^{\mu} + S^{\mu}_{\ \ i} \right) \right) - \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
+ E a_{i} - S^{j} K_{ji}
\end{align*}

\begin{align*}
0 \ &= \
-\frac{1}{\sqrt{-g}} \, \partial_{\mu} \left( \sqrt{-g} \left( E \, n^{\mu} + S^{\mu} \right) \right) - S^{\nu} a_{\nu} + S^{jk} K_{jk}
\end{align*}

\begin{align*}
\partial_{t} \left( \sqrt{-g} ( S_{i} n^{t} + S^{t}_{\ \ i} ) \right)
\ + \
\partial_{j} \left( \sqrt{-g} ( S_{i} n^{j} + S^{j}_{\ \ i} ) \right)
\ = \ \sqrt{-g} \left( \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
- E a_{i} + S^{j} K_{ji} \right)
\end{align*}

\begin{align*}
\partial_{t} \left( \sqrt{-g} ( E \, n^{t} + S^{t} ) \right)
\ + \
\partial_{i} \left( \sqrt{-g} ( E \, n^{i} + S^{i} ) \right)
\ = \ \sqrt{-g} \left( - S^{\nu} a_{\nu} + S^{jk} K_{jk} \right)
\end{align*}

Using that, $n^{\mu} = (1/\alpha, -\beta^{i}/\alpha)$, $\sqrt{-g} = \alpha \sqrt{\gamma}$, and, $a_{\mu} = D_{\mu} \log \alpha$, where, $D_{\mu} \equiv \perp^{\nu}_{\ \ \mu} \nabla_{\nu}$, we find,

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} S_{i} \right)
\ + \
\partial_{j} \left( \sqrt{\gamma} (  \alpha S^{j}_{\ \ i} - S_{i} \beta^{j}) \right)
\ = \ \sqrt{\gamma}\left( \alpha \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
 + \alpha S^{j} K_{ji} - E \partial_{i} \alpha \right)
\end{align*}

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} E  \right)
\ + \
\partial_{i} \left( \sqrt{\gamma} (\alpha S^{i} - E \, \beta^{i} ) \right)
\ = \ \sqrt{\gamma} \left(\alpha S^{jk} K_{jk} - S^{\nu} \partial_{\nu} \alpha \right)
\end{align*}

Using the definitions that, $S^{i} = (E+p)v^{i}$, and, $S^{ij} = S^{i} v^{j} + p \gamma^{ij}$, in a perfect fluid,

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} S_{i} \right)
\ + \
\partial_{j} \left( \sqrt{\gamma} (  \alpha (S_{i}v^{j} + p \delta^{j}_{\ \ i}) - S_{i} \beta^{j}) \right)
\ = \ \sqrt{\gamma} \left(\alpha \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
+ \alpha S^{j} K_{ji} - E \partial_{i} \alpha \right)
\end{align*}

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} E \right)
\ + \
\partial_{i} \left( \sqrt{\gamma} (\alpha (E+p)v^{i} - E \, \beta^{i} ) \right)
\ = \ \sqrt{\gamma} \left(\alpha S^{jk} K_{jk} - S^{\nu} \partial_{\nu} \alpha \right)
\end{align*}

And rearranging the terms yields,

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} S_{i} \right)
\ + \
\partial_{j} \left( \sqrt{\gamma} (  S_{i} (\alpha v^{j} - \beta^{j}) + \alpha p \delta^{j}_{\ \ i} \right)
\ = \ \sqrt{\gamma} \left(\alpha \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right)
+ \alpha S^{j} K_{ji} - E \partial_{i} \alpha \right)
\end{align*}

\begin{align*}
\partial_{t} \left( \sqrt{\gamma} E \right)
\ + \
\partial_{i} \left( \sqrt{\gamma} \left( E (\alpha v^{i} - \beta^{i}) + \alpha p v^{i} \right) \right)
\ = \ \sqrt{\gamma} \left(\alpha S^{jk} K_{jk} - S^{\nu} \partial_{\nu} \alpha \right)
\end{align*}


Note that,

\begin{align*}
\alpha S^{\rho} \left( \Gamma_{\rho \mu i} n^{\mu} + K_{\rho i} \right)
\ &= \
\alpha S^{j} \left( \Gamma_{j \mu i} n^{\mu} + K_{ij} \right) \\
\ &= \
S^{j} \left( \Gamma_{j t i} - \Gamma_{j k i} \beta^{k} + \alpha K_{ij} \right) \\
\ &= \ 
S^{j} \left( \frac{1}{2}( \partial_{t} \gamma_{ij} + \partial_{i} \beta_{j} - \partial_{j}\beta_{i} ) - \Gamma_{j k i} \beta^{k} + \alpha K_{ij} \right) \\
\ &= \ 
S^{j} \left( \partial_{i} \beta_{j} - \partial_{i} \gamma_{jk} \beta^{k} \right) \\
\ &= \
S^{j} \gamma_{jk} \partial_{i} \beta^{k} \\
\ &= \
S_{k} \partial_{i} \beta^{k}
\end{align*}
where we used that,
\begin{align*}
\frac{1}{2} \, \partial_{t} \gamma_{ij} + \alpha K_{ij}
\ &= \
D_{i} \beta_{j} + D_{j} \beta_{i} \\
\ &= \
\frac{1}{2} \left( \partial_{i} \beta_{j} + \partial_{j} \beta_{i} \right) - \Gamma_{kij} \beta^{k}
\end{align*}
Furthermore, we use that,
\begin{align*}
\Gamma_{\rho \mu i} S^{\mu \rho} \ = \ \frac{1}{2} \, S^{jk} \partial_{i} g_{jk}
\end{align*}

\begin{align*}
\alpha \Gamma_{\rho \mu i} \left(S^{\rho} n^{\mu} + S^{\mu \rho} \right) + \alpha S^{j} K_{ji} - E \partial_{i} \alpha
\ &= \
S_{k} \partial_{i} \beta^{k} + \frac{1}{2} \, \alpha S^{jk} \partial_{i} g_{jk} - E \partial_{i} \alpha
\end{align*}


The conservation equations can be written as a first-order flux-conservative hyperbolic syetem of equations.
These conservation equations are, however, defined in terms of five **conservative variables**, $\boldsymbol{C}\equiv (D, S_{i}, \tau)$, defined as,

\begin{align*}
    D
    \ &\equiv \
    \rho W  \\
    S_{i}
    \ &\equiv \
    \rho h W^{2} v_{i} \ = \ (E+p)v_{i} \\
    \tau
    \ &\equiv \
    \rho h W^{2} - p - D \ = \ E - D
\end{align*}

The first-order hyperbolic system of equations then reads,
\begin{align*}
    \partial_{t} \left(\sqrt{\gamma} \, \boldsymbol{C} \right)
    \ + \
    \partial_{i} \left(\sqrt{\gamma} \, \boldsymbol{F}^{i} \right)
    \ = \
    \sqrt{\gamma} \, \boldsymbol{S}
\end{align*}
with fluxes, $\boldsymbol{F}^{i}$, and source term, $\boldsymbol{S}$, defined as, 
\begin{align*}
    \boldsymbol{C}
    \ &\equiv \
    \left(
    \begin{matrix}
    \rho W  \\
    \rho h W^{2} v_{j} \\
    \rho h W^{2} - p - D 
    \end{matrix}
    \right) \\
    \boldsymbol{F}^{i}
    \ &\equiv \
    \left(
    \begin{matrix}
         D     \left(\alpha v^{i} - \beta^{i}\right) \\
         S_{j} \left(\alpha v^{i} - \beta^{i}\right) + \alpha p \delta^{i}_{j} \\
         \tau  \left(\alpha v^{i} - \beta^{i}\right) + \alpha p v^{i}
    \end{matrix}
    \right) \\
    \boldsymbol{S}
    \ &\equiv \
    \left(
    \begin{matrix}
         0 \\
         \frac{1}{2} \, \alpha S^{ik} \partial_{j} \gamma_{ik} + S_{i} \partial_{j} \beta^{i} - E \partial_{j} \alpha \\
         \alpha S^{ik} K_{ik} - S^{i} \partial_{i} \alpha
    \end{matrix}
    \right) .
\end{align*}

In the xCFC scheme the extrinsic curvature can be approximated as,
\begin{align*}
K_{ij} \ = \ \psi^{-10} \mathsf{A}_{ij}
\end{align*}
in which $\nabla_{i}$ is the covariant derivative associated with $\gamma_{ij}$.

\begin{align*}
S^{ij} K_{ij} \ = \ \rho h W^{2} v_{r}^{2} \psi^{-10} \mathsf{A}_{rr}
\end{align*}

\begin{align*}
S^{ij} \partial_{r} \gamma_{ij}
\ &= \
4 \rho h W^{2} v_{r}^{2} \partial_{r} \log \psi + p \gamma^{ik} \partial_{j} \gamma_{ik} \\
\ &= \
4 \rho h W^{2} v_{r}^{2} \partial_{r} \log \psi + p \left( 12 \partial_{r} \log \psi + \frac{4}{r} \right) \\
\ &= \ 
4 \left(\rho h W^{2} v_{r}^{2} + 3 p\right) \partial_{r} \log \psi + \frac{4p}{r}  \\
\end{align*}

$S \equiv \gamma^{ij} S_{ij} = \rho h W^{2} v^{i} v_{i} + 3 p$

$\gamma_{ij} = \psi^{4} f_{ij}$

So note that in spherical symmetry, $S_{r} = \psi^{4} S^{r}$.

\begin{align*}
    \boldsymbol{S}
    \ &\equiv \
    \left(
    \begin{matrix}
         0 \\
         2 \alpha S \partial_{r} \log \psi + \frac{2\alpha p}{r} + S_{i} \partial_{j} \beta^{i} - E \partial_{j} \alpha \\
         \alpha \rho h W^{2} v_{r}^{2} \psi^{-10} \mathsf{A}_{rr} - S^{i} \partial_{i} \alpha
    \end{matrix}
    \right) .
\end{align*}

```python
import numpy             as np
import matplotlib.pyplot as plt

from scipy.integrate import solve_ivp
from ipywidgets      import interact


class Hydro:

    def __init__(N, R):

        self.rs = np.linspace(0, R, N)   # radial grid

        self.rho = np.zeros(N)           # density
        self.v_r = np.zeros(N)           # radial velocity
        self.eps = np.zeros(N)           # specific internal energy


    def eos_ideal_fluid(self, rho, eps, gamma):
        '''
        Ideal fluid equation of state.
        '''
        p = (gamma - 1) * self.rho * self.eps
        return p
    
    def eos_polytropic(self, rho, K, gamma):
        '''
        Polytropic equation of state.
        '''
        p = K * rho**gamma
        return p

    def h(self):
        '''
        Enthalpy.
        '''
        return 1.0 + self.eps + self.p / self.rho
```

```python
def rhs(t, y, r, alpha, beta_r, psi):
    """
    Right-hand side of the conservation equations.
    """

    N = len(y) // 3

    # Unpack conserved variables
    tilde_D   = y[   :  N]
    tilde_S_r = y[  N:2*N]
    tilde_tau = y[2*N:3*N]

    # Define helper variables
    psi4  = psi**4
    psi6  = psi**6
    psi10 = psi**10
    log_r = np.log(r)
    inv_r    = 1.0 / r
    inv_psi6 = 1.0 / psi6

    # Extract sqrt of the (spatial) metric determinant, i.e. sqrt(det(gamma))
    sqrt_gamma = psi6 * r**2

    # Conserved variables
    D   = tilde_D   * inv_psi6
    S_r = tilde_S_r * inv_psi6
    tau = tilde_tau * inv_psi6

    # Convert S_r to S^r
    S_r /= psi4

    # print(np.min(np.abs(D)))
    # print(np.min(np.abs(S_r)))
    # print(np.min(np.abs(tau))

    # Convert conserved to primitive variables
    (rho, p, v_r, W, h) = cons2prim(D=D, S_r=S_r, tau=tau, eos=eos)

    # Variables required by metric solver
    a   = rho * h * W**2
    E   = a - p
    S_r = a * v_r
    S   = S_r * v_r + 3.0 * p

    # Rescaled variables required by metric solver
    tilde_E   = psi6 * E
    tilde_S_r = psi6 * S_r
    tilde_S   = psi6 * S

    # Solve for metric 
    alpha[:], beta_r[:], psi[:], A_rr = solve_metric(tilde_E, tilde_S_r, tilde_S)

    # Get metric derivatives
    d_alpha  = np.gradient(alpha,  log_r) * inv_r
    d_beta_r = np.gradient(beta_r, log_r) * inv_r
    d_psi    = np.gradient(psi   , log_r) * inv_r
    
    # Flux variables
    b     = alpha * v_r - beta_r
    F_D   = D          * b
    F_r_S = S_r * psi4 * b + alpha * p
    F_tau = tau        * b + alpha * p * v_r

    # Source variables
    S_r_S = 2 * alpha * (S * d_psi / psi + p * inv_r) + S_r * psi4 * d_beta_r - E * d_alpha
    S_tau =     alpha * rho * h * W**2 * v_r**2 / psi10 * A_rr - S_r * d_alpha

    # print('F', sqrt_gamma * F_r_S)
    # print('S', S_r_S)
    # print('FLUX', -np.gradient(sqrt_gamma * F_r_S, log_r) * inv_r)

    # Right-hand sides of conservation equations
    dC_D_dt   = -np.gradient(sqrt_gamma * F_D,   log_r) * inv_r**3
    dC_S_r_dt = -np.gradient(sqrt_gamma * F_r_S, log_r) * inv_r**3 + psi6 * S_r_S
    dC_tau_dt = -np.gradient(sqrt_gamma * F_tau, log_r) * inv_r**3 + psi6 * S_tau
    
    # print(np.min(np.abs(dC_D_dt)))
    # print(np.min(np.abs(dC_S_r_dt)))
    # print(np.min(np.abs(dC_tau_dt)))

    # print('dC_S_r_dt', dC_S_r_dt)

    # # plt.plot(r, np.abs(+dC_S_r_dt))
    # # plt.plot(r, np.abs(-dC_S_r_dt))
    # # plt.plot(r, dC_S_r_dt)
    # plt.plot(r, np.gradient(sqrt_gamma * F_r_S, log_r) * inv_r**3)
    # plt.plot(r, psi6 * S_r_S)
    # plt.yscale('log')
    # plt.xscale('log')
    # plt.show()
    
    # Return the derivatives
    return np.concatenate((dC_D_dt, dC_S_r_dt, dC_tau_dt))


t_fin = 1.0e-2

# Define initial conditions

# Helper variables
W_0 = 1.0 / np.sqrt(1.0 - v_r_0**2)
h_0 = 1.0 + eps_0 + p_0 / rho_0

# Conserved variables
D_0   = rho_0 * W_0
S_r_0 = rho_0 * h_0 * W_0**2 * v_r_0
tau_0 = rho_0 * h_0 * W_0**2 - p_0 - D_0

# Sqrt of (spatial) metric determinant, i.e. sqrt(det(gamma))
sqrt_gamma_0 = psi_0**6 * R_0**2

# Rescaled conserved variables
tilde_D_0    = psi_0**6 * D_0
tilde_S_r_0  = psi_0**6 * S_r_0
tilde_tau_0  = psi_0**6 * tau_0

y0 = np.concatenate((tilde_D_0, tilde_S_r_0, tilde_tau_0))


R      = R_0     .copy()
alpha  = alpha_0 .copy()
beta_r = beta_r_0.copy()
psi    = psi_0   .copy()


sol = solve_ivp(
    fun          = rhs,
    t_span       = [0.0, t_fin],
    y0           = y0,
    method       = 'RK45',
    rtol         = 1.0e-12,
    atol         = 1.0e-16,
    args         = (R, alpha, beta_r, psi),
)
```

```python
class Metric:

    @property
    def alpha(self):
        return self._alpha
    
    @alpha.setter
    def alpha(self, value):
        self._alpha = value

    @property
    def psi(self):
        return self._psi
    
    @psi.setter
    def psi(self, value):
        self._psi = value



metric = Metric()
metric.alpha = 0.0
metric.psi   = 0.0

def rhs(t, y, metric):

    metric.alpha += 1.0

    return y + metric.alpha

y0 = np.array([1.0])


sol = solve_ivp(
    fun          = rhs,
    t_span       = [0.0, 10.0],
    y0           = y0,
    method       = 'BDF',
    rtol         = 1.0e-12,
    atol         = 1.0e-16,
    args         = (metric,),
)
```
