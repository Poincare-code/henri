```python
import numpy             as np
import matplotlib.pyplot as plt

plt.figure(dpi=1000)

t = np.linspace(-np.pi,3*np.pi, 1000)

def amp(t):
    A = 0.025*np.exp(0.7*t) + 0.3
    A[t>6] = 2 * np.exp(4*(6-t[t>6]))
    return A

def frq(t):
    F = np.exp(0.75*t) * 0.07 + 10
    F[t>6] = np.max(F[t<6]) * 3.9
    return F

signal = lambda t: amp(t) * np.sin(frq(t) * t)
signal2 = lambda t: amp(t) * np.sin(12 * t)


# plt.plot(t, amp(t))
# plt.plot(t, frq(t))
plt.plot(t, signal(t), c='k')
# plt.fill_between(t, signal(t)-0.1, signal(t)+0.1, color='k', alpha=0.3, linewidth=0)
# plt.plot(t, signal(t), c='w')
# plt.plot(t, signal2(t))

plt.savefig('signal_w.png', transparent=True)
```
